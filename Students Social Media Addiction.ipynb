## Doing a T test for Undergrades, Grades, and HighSchool to learn the usage pattern

```python
from scipy import stats as st
grad_scores = df_cleaned[df_cleaned['Academic_Level'] == 'Graduate']['Addicted_Score']
ugrad_scores = df_cleaned[df_cleaned['Academic_Level'] == 'Undergraduate']['Addicted_Score']

t_stat, p_value = st.ttest_ind(grad_scores, ugrad_scores, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

#### The p value is less than 0.05 but at 5% level so it is true that udergraduate are more addicted but the - value of T stats says that Grads are comparetively less addicted compared to undergrads

```python
high_school = df_cleaned[df_cleaned['Academic_Level']=='High School']['Addicted_Score']
t_stat,p_value=st.ttest_ind(high_school,grad_scores, equal_var=False)
print('T-statstics:', t_stat)
print('P-Value:',p_value)
```

#### we can see that the p values is way smaller than 0.05 and hence we can confirm the null hypothesis that hgih schoolstudents are more addicted to social media

```python
t_stat,p_value=st.ttest_ind(high_school,ugrad_scores, equal_var=False)
print('T-statstics:', t_stat)
print('P-Value:',p_value)
```

#### As we can see again the Tstat is way higher and the pvalue is still smaller than 0.05 so again we can conclude that the nullhypothesis is true and high school student are more addicted to social media than undergraduates

## Trying to Classify relationship conflicts depending upon user pattern and platform.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

lb=LabelEncoder()
df_cleaned['Most_Used_Platform_trans']=lb.fit_transform(df_cleaned['Most_Used_Platform'])

X=df_cleaned[['Most_Used_Platform_trans','Avg_Daily_Usage_Hours','Sleep_Hours_Per_Night','Addicted_Score']]
y=df_cleaned['Conflicts_Over_Social_Media']

sc=StandardScaler()
X_sc=sc.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_sc, y, test_size=0.2, random_state=42)
```

```python
model=LogisticRegression(multi_class='multinomial', solver='lbfgs',max_iter=500)
model.fit(X_train,y_train)

y_prediction=model.predict(X_test)

print("Accuracy:", accuracy_score(y_prediction,y_test))

print(classification_report(y_prediction,y_test))
```

### No we can see that the model has captured the data and made an accuracy of 92 in predicting the relationship conflict depending upon the usage behaviour
### with an impressive f1-score of 97 , 97 , 91 for all the major classes and 63 for the minor class

## Clustering the data to find a pattern within 

```python
from sklearn.cluster import KMeans
X=df_cleaned[['Mental_Health_Score','Avg_Daily_Usage_Hours','Sleep_Hours_Per_Night','Addicted_Score','Conflicts_Over_Social_Media']]
X_sc=sc.fit_transform(X)

# determining the K using elbow
winthincSquaredSum=[]
k=10
for i in range(1,k+1):
    kmeans=KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(X_sc)
    winthincSquaredSum.append(kmeans.inertia_)

# plotting the graph

plt.figure(figure=(5,5))
ax=sns.lineplot(x= range(1,k+1), y=winthincSquaredSum, marker='o' , linestyle='--')
```

#### from the above elbow method we can understand that 3 clusters will be perfect as the slops has drastic changes from 1 to and then 2 to 3 after that it starts tosmoothen the fall.
         so using 3 clusters will  perfectly  segregate the data into meaning full clusters

```python
features_for_clustering = [
    'Addicted_Score',
    'Avg_Daily_Usage_Hours',
    'Sleep_Hours_Per_Night',
    'Mental_Health_Score',
    'Conflicts_Over_Social_Media'
]
k=3
kmeans=KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
df_cleaned['Clusters']= kmeans.fit_predict(X_sc)
print(f"\n--- K-Means Clustering with k = {k} ---")
```
